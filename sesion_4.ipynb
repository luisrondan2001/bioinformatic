```python
from pydna.genbank import Genbank
```

```python
gb = Genbank("myself@mail.com")
YEp21 = gb.nucleotide("KC562906")
print(YEp21)
```

```python
import dna_features_viewer
from dna_features_viewer import BiopythonTranslator
```

```python
from dna_features_viewer.CircularGraphicRecord.CircularGraphicRecord import CircularGraphicRecord
bio = BiopythonTranslator()
graphic = bio.translate_record(YEp21, record_class = CircularGraphicRecord)
graphic.plot()
```

```python
from Bio.Restriction import SmaI
plinear = YEp21.linearize(SmaI)
bio = BiopythonTranslator()
graphic = bio.translate_record(plinear)
graphic.plot()
```

```python
pcircular = plinear.looped()
print(pcircular)
bio = BiopythonTranslator()
graphic = bio.translate_record(pcircular, record_class = CircularGraphicRecord)
graphic.plot()
```

```python
from Bio.Restriction import EcoRI
fa = YEp21.cut(EcoRI)[0]
fb = YEp21.cut(EcoRI)[1]
graphic = BiopythonTranslator().translate_record(fa)
graphic.plot()
graphic = BiopythonTranslator().translate_record(fb)
graphic.plot()
```

```python
from pydna.parsers import parse
from pydna.amplify  import pcr
```

```python
primer_r , primer_f = parse (''' >klLAC_rv
                                  ttaaacagattctgcctctg
                                 >klLAC_fw
                                  aaatggcagatcattcgag
                                 ''', ds = False)
print (primer_f)
print (primer_r)
```

```python
gb = Genbank("myself@mail.com")
gen = gb.nucleotide("X06997")
print(gen)
```

```python
PCR  = pcr(primer_f, primer_r, gen)
print(PCR)
print(PCR.features)
print(PCR.seq)
```

```python
print(PCR.figure())
```

```python
print(PCR.program())
```

```python
from pydna.gel import gel
from pydna.ladders import PennStateLadder
```

```python
gel([PennStateLadder, [PCR]])
```

```python
ADN_rec = (plinear + PCR).looped()
print(ADN_rec )
graphic = BiopythonTranslator().translate_record(gen)
graphic.plot()
graphic = BiopythonTranslator().translate_record(PCR)
graphic.plot()
```
