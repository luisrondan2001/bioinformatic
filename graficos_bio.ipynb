##Diagrama de cuerdas

Un diagrama de cuerdas es una visualización que muestra las relaciones entre diferentes elementos. En este tipo de diagrama, cada elemento se representa como un arco en un círculo, y las relaciones entre los elementos se representan como cuerdas que conectan los arcos correspondientes. El grosor de las cuerdas puede representar la fuerza de la relación, y los colores pueden utilizarse para codificar diferentes categorías o grupos de elementos. Los diagramas de cuerdas se utilizan comúnmente en campos como la biología, la sociología y la visualización de datos
##Heatmap

Un heatmap, o mapa de calor, es una representación visual de datos que utiliza colores para indicar la magnitud de los valores en una matriz. Por lo general, se utiliza en el contexto de análisis de datos y visualización para resaltar patrones, tendencias o variaciones en los datos. Cada celda de la matriz se colorea de acuerdo con el valor que representa, lo que permite una rápida identificación de las áreas con valores altos o bajos. Los mapas de calor son comúnmente utilizados en campos como la biología, la meteorología, las finanzas y el análisis de datos para visualizar la densidad, la correlación, la distribución o la agrupación de los datos
